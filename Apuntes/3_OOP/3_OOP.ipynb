Veamos en la sesin anterior que cuando creamos un objeto de la clase *Coche* sin ms, este objeto adquiere los atributos por defecto de la clase. Si queremos definirlo bien "de primeras" (en su instanciacin o creacin) para diferenciarlo de otros coches, necesitamos algo ms.  

Ese algo ms que me permite la definicin inicial, es lo que se denomina constructor de la clase.

Ser un mtodo, muy similar a una funcin, al que daremos unos argumentos de entrada que sustituirn a los valores por defecto y que nos permitirn diferenciar esa instancia de otras instancias de la misma clase.

**Cmo se define un constructor?** Mediante la funcin `__init__`, dentro de la clase.

````python
class Coche:
    puertas = 4
    ruedas = 4

    def __init__(self, marca_coche):
        self.marca = marca_coche
````

````python
tesla = Coche()
````

````python
tesla = Coche("Tesla")
print(tesla.marca)
````

En la declaracin del constructor hemos metido la palabra `self`. **Lo tendremos que poner siempre**. Hace referencia a la propia instancia de coche, es decir, a cuando creemos coches nuevos.

En este caso estamos diferenciando los atributos comunes de la clase *Coche*, de los atributos particulares de los coches, como por ejemplo, la marca. Por eso la marca va junto con `self`, porque no hace referencia a la clase genrcia de coche, sino a cada coche que creemos.

Ahora ya podemos diferenciar los coches por su marca. Para acceder al atributo de la marca, lo hacemos igual que con los anteriores.

````python
mercedes = Coche("Mercedes")
renault = Coche("Renault")
print(mercedes.ruedas)
print(renault.ruedas)
````

##  Mtodos

### Introduccin

Son funciones que podemos definir dentro de las clases. Estas funciones cambiarn el estado de algn atributo o realizarn clculos que nos sirvan de output. Un ejemplo sencillo puede ser, un mtodo de la clase coche que saque la potencia en kilovatios, en vez de en caballos. O si tiene un estado de mantenimiento (ITV pasada o no), que modifique ese estado.

El constructor es un tipo de mtodo. La diferencia con el resto de mtodos radica en su nombre, `__init__`. La sintaxis para definir los mtodos es como si fuese una funcin. Y luego para llamar al mtodo se utiliza `objeto.metodo(argumentos_metodo)`. Esto ya lo hemos usado anteriormente, cuando haciamos un `string.lower()`, simplemente llambamos al mtodo `lower()`, que no requera de argumentos, de la clase *string*.

````python
class Coche:
    ruedas = 4

    def __init__(self, marca_coche, num_puertas = 4):
        self.marca = marca_coche
        self.puertas = num_puertas

    def caracteristicas(self):
        return "Marca:" + self.marca + ", Puertas:" + str(self.puertas)
````

````python
ford_ka = Coche("Ford")
ford_ka.caracteristicas()
````

Fjate que para llamar a las ruedas se usa `self`, a pesar de que no lo habamos metido en el constructor. As evitamos llamar a otra variable del programa que se llame *ruedas*. Nos aseguramos que son las ruedas de ese coche con el `self`.

````python
class Coche:
    ruedas = 4

    def __init__(self, marca_coche, precio, num_puertas = 4,):
        self.marca = marca_coche
        self.puertas = num_puertas
        self.precio = precio

    def caracteristicas(self):
        return "Marca:" + self.marca + ", Puertas:" + str(self.puertas)

    def precio_actual(self, agnos):
        if agnos < 5:
            return self.precio*0.7
        elif agnos > 5 and agnos < 10:
            return self.precio * 0.5
        else:
            return self.precio * 0.3
````

````python
jeep_cherokee = Coche("Jeep", 3500)
jeep_cherokee.precio_actual(8)
````
