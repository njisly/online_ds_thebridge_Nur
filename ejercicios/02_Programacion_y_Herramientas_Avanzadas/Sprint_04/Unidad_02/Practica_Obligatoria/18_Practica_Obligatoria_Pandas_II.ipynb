1. Clasifiquemos los vuelos en Larga Distancia (LD), Media Distancia (MD) y Regionales (R). Con los siguientes lmites: ms de 9000km, entre 1500 y 9000 kms y resto de vuelos. Aade una columna al `DataFrame` que tenga esa clasificacin por vuelo.

```python
def clasificador(row):
    if row["distancia"] < 9000:
        clasificacion = "LD"
    elif row["distancia"] >= 9000 and distancia >= 1500:
        clasificacion = "MD"
    else:
        clasificacion = "R"
    return clasificacion

df_viajes_aereos["Distancia"].apply(clasificador)

df_viajes_aereos["Categoria"] = df_viajes_aereos["Distancia"].apply(clasificador)
```

2. Crea ahora una columna que recoja el ndice de contaminacin de cada viaje calculado como el consumo de keroseno por kilometro recorrdio. Llama IC a esa columna.

```python
def ic(row):
    distancia = row["Distancia"]
    consumo_kg = row["consumo_kg"]
    consumo_km = consumo_kg/distancia
    return consumo_km

df_viajes_aereos["IC"] = df_viajes_aereos[["Distancia", "consumo_kg"]].apply(ic, axis=1)
```
